```python
wav_file = './Dataset/Lung/lung_01.wav'  # ðŸ‘ˆ change if needed
target_fs = 16000
```

```python
import numpy as np, soundfile as sf, librosa, csv
from scipy.signal import butter, filtfilt, stft, istft, hilbert
import matplotlib.pyplot as plt

# --- Read and resample ---
x, fs = sf.read(wav_file)
if x.ndim > 1:
    x = np.mean(x, axis=1)
if fs != target_fs:
    x = librosa.resample(y=x, orig_sr=fs, target_sr=target_fs)
    fs = target_fs
x = x / (np.max(np.abs(x)) + 1e-12)

# --- Bandpass 100â€“2000 Hz ---
b, a = butter(4, [100/(fs/2), 2000/(fs/2)], btype='band')
x_bp = filtfilt(b, a, x)

# --- Spectral subtraction noise reduction ---
n_fft = 2048
hop = 256
win = np.hamming(512)
f, t, Zxx = stft(x_bp, fs=fs, nperseg=len(win), noverlap=len(win)-hop)
mag = np.abs(Zxx)
pow_spec = mag**2
noise_spec = np.mean(pow_spec[:, :6], axis=1, keepdims=True)
clean_pow = np.maximum(pow_spec - 2*noise_spec, 1e-12)
gain = np.sqrt(clean_pow / (pow_spec + 1e-12))
Zxx_clean = Zxx * gain
_, y = istft(Zxx_clean, fs=fs, window=win, nperseg=len(win), noverlap=len(win)-hop)

# --- Envelope + simple cough detection ---
analytic = hilbert(y)
env = np.abs(analytic)
thr = np.mean(env) + 2*np.std(env)
events = env > thr

# --- MFCC extraction (fixed Librosa syntax) ---
mel = librosa.feature.melspectrogram(
    y=y,
    sr=fs,
    n_fft=n_fft,
    hop_length=hop,
    n_mels=40,
    fmin=50,
    fmax=3000,
    power=2.0
)
mfccs = librosa.feature.mfcc(
    S=librosa.power_to_db(mel, ref=np.max),
    n_mfcc=13
)
mfcc_mean = np.mean(mfccs, axis=1)
mfcc_std = np.std(mfccs, axis=1)
feat_row = np.concatenate([mfcc_mean, mfcc_std])
np.savetxt('mfccs_python.csv', mfccs.T, delimiter=',')
np.savetxt('features_python.csv', feat_row.reshape(1,-1), delimiter=',')

# --- Plots ---
plt.figure(figsize=(12,8))
plt.subplot(3,1,1)
plt.plot(x_bp)
plt.title("Bandpassed Lung Sound (100â€“2000 Hz)")
plt.subplot(3,1,2)
plt.plot(env)
plt.axhline(thr, color='r', linestyle='--')
plt.title("Envelope with Detected Events (Threshold)")
plt.subplot(3,1,3)
librosa.display.specshow(
    librosa.power_to_db(mel, ref=np.max),
    sr=fs,
    hop_length=hop,
    x_axis='time',
    y_axis='mel'
)
plt.title("Mel-Spectrogram")
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()
plt.show()

print('âœ… Done. Saved features_python.csv and mfccs_python.csv')
```
